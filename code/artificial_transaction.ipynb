# Generates Artificial Transactions

This is so that I don't expose my own personal finances

Cost - range per payment first year (e.g. gas & fuel range is 70;100)
Increase - % per year increase (e.g. gas & fuel is ... 5)
Jump - year;amount jumped (e.g. gas & fuel is ... 2020;-40;;2021;100)
DateDiff - min;max date difference between the payments (days) (e.g. gas & fuel ... 3;15)
Vendors - list of vendor names can be bought from (e.g. Mac's, circle k)

```python
import pandas as pd
import random
from datetime import datetime

df_cat = pd.read_csv(r'/workspaces/PersonalFinanceDashboard/resources/CategoryGrouping.csv')
```

```python
# initialise data of lists.
col = ['Date', 'Description', 'Original Description', 'Amount', 'Transaction Type', 'Category', 'Account Name', 'Labels', 'Notes']
 
# Create DataFrame
df_trans = pd.DataFrame(columns=col)
```

```python
years = [2018, 2019, 2020, 2021, 2022]

for year in years:
    for row in range(len(df_cat['Cost'].dropna())):
        date = random.randint(int(df_cat['DateDiff'][row].split(';')[0]),int(df_cat['DateDiff'][row].split(';')[1]))
        while date <= 365:
            # print(datetime.strptime(str(year) + "-" + str(date), "%Y-%j").strftime("%m/%d/%Y"))
            # print('\t' + str(random.randint(int(df_cat['Cost'][0].split(';')[0]),int(df_cat['Cost'][0].split(';')[1]))+(0.01 * random.randint(0,100))))
            description = random.choice(df_cat['Vendors'][row].split(';'))

            df_trans = df_trans.append({'Date':datetime.strptime(str(year) + "-" + str(date), "%Y-%j").strftime("%m/%d/%Y"), 
                                        'Description':description,
                                        'Original Description':description,
                                        'Amount':str(round(random.randint(int(df_cat['Cost'][row].split(';')[0]),int(df_cat['Cost'][row].split(';')[1]))+(0.01 * random.randint(0,100)), 2)),
                                        'Transaction Type':'debit',
                                        'Category':df_cat['Category'][row],
                                        'Account Name':'Visa',
                                        'Labels':'',
                                        'Notes':''}, 
                                        ignore_index=True)

            date += random.randint(int(df_cat['DateDiff'][row].split(';')[0]),int(df_cat['DateDiff'][row].split(';')[1]))
```

```python
df_trans.to_csv(r'/workspaces/PersonalFinanceDashboard/resources/faketransactions.csv', index=False)
```
